```python
def findInverse(n,e):
    r1 = e
    r2 = n
    t1 = 0
    t2 = 1

    while(r2 > 0):
        q = r1//r2
        r = r1 - q*r2
        r1 = r2
        r2 = r
        t = t1 - q*t2
        t1 = t2
        t2 = t

    if(r1 == 1):
        return t1


def CRT(a,b,p,q):

    M = p*q
    M1 = q
    M2 = p
    m1 = p
    m2 = q

    M1I = findInverse(M1,m1)
    M2I = findInverse(M2,m2)

    return (a*M1*M1I + b*M2*M2I)%M



p, q = list(map(int, input().split()))
n = p*q

P = int(input())
while(P >= n):
    print("Enter message less than n")
    P = int(input())

C = (P**2)%n

print("Encrypted Message: ",C)

print('Decrypting...............')

a1 = (C**((p+1)//4))%p
a2 = -a1
b1 = (C**((q+1)//4))%q
b2 = -b1

P1 = CRT(a1,b1,p,q)
P2 = CRT(a1,b2,p,q)
P3 = CRT(a2,b1,p,q)
P4 = CRT(a2,b2,p,q)

print(P1,P2,P3,P4)
# assert (P1==P or P2==P or P3==P or P4 == P)
```
